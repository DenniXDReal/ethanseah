```python
#Ethan
import pandas as pd
import sys
#input for wanted year perod and region
period = input("Enter year period (1: 1978-1987, 2: 1988-1997, 3: 1998-2007, 4: 2008-2017): ")
region = input("Enter region (1: Asia, 2: Europe, 3: Others): ")

#if the field is empty then both period and region input are required...exiting will be printed
if not period or not region:
    print("Both period and region inputs are required. Exiting...")
    sys.exit(0)

#self,period,region and path are the parameters,self.period,region and path is the parameters
class VisitorsAnalyticsUtils:
    def __init__(self, period, region, path):
        self.period = period
        self.region = region
        self.path = path
    
    #loadDataFiles reads the csv and turns it into a dataframe
    def loadDataFile(self):
        return pd.read_csv(self.path)
    
    #parsed data filters the data into specific input the users want
    def parseDataFile(self):
        data = self.loadDataFile()
        data['Year'] = data.iloc[:, 0].str.split().str[0]
        data = data.dropna()
        
        #List for regions
        asia = [" Brunei Darussalam ", " Indonesia ", " Malaysia ", " Philippines ", " Thailand ", " Viet Nam ", " Myanmar ", " Japan ", " Hong Kong ", " China ", " Taiwan " ," Korea, Republic Of ", " India ", " Pakistan ", " Sri Lanka ", " Saudi Arabia ", " Kuwait ", " UAE "]
        europe = [' United Kingdom ', ' Germany ', ' France ', ' Italy ', ' Netherlands ', ' Greece ', ' Belgium & Luxembourg ', ' Switzerland ', ' Austria ', ' Scandinavia ', ' CIS & Eastern Europe '  ]
        others = [' USA ', ' Canada ', ' Australia ', ' New Zealand ', ' Africa ']

        #getting the data greater than and equal to and less than and equal to for every 10 years
        if self.period == '1':
            period_data = data.loc[(data['Year'] >= '1978') & (data['Year'] <= '1987')]
        elif self.period == '2':
            period_data = data.loc[(data['Year'] >= '1988') & (data['Year'] <= '1997')]
        elif self.period == '3':
            period_data = data.loc[(data['Year'] >= '1998') & (data['Year'] <= '2007')]
        elif self.period == '4':
            period_data = data.loc[(data['Year'] >= '2008') & (data['Year'] <= '2017')]
        else:
            print("Period is not in range. Exiting...")
            sys.exit(0)

        period_data = period_data.drop('Year', axis=1)

        #if a region is selected the period chosen will copy
        if self.region == '1':
            region_data = period_data[asia].copy()
        elif self.region == '2':
            region_data = period_data[europe].copy()
        elif self.region == '3':
            region_data = period_data[others].copy()
        else:
            print("Region is not in range. Exiting...")
            sys.exit(0)
        
        return region_data
    
    #totalling up all the data and return top to show the countries with the largest sum
    def getTop3Countries(self):
        rawdata = self.parseDataFile()
        data = rawdata.T 
        data = data.apply(lambda x: pd.to_numeric(x.str.replace(',', ''), errors='coerce').fillna(0).astype('int64'))
        row_sums = data.sum(axis=1, skipna=True)
        top = row_sums.nlargest(3)
        top = top.astype('int64') 
        return top


#printing out the top 3 countries
data = VisitorsAnalyticsUtils(period, region, './Int_Monthly_Visitor.csv')
print("***** TOP 3 COUNTRIES *****")
print(data.getTop3Countries())

```
